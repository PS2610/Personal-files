### **1. Which EEG amplifier was used for the experiment?**

- The experiment utilized the g.HIamp amplifier system in a 128-channel configuration. Specifically, the system was configured as a "128+16" version, which consists of two 64-channel connection boxes plus one 16-channel connector box. This high-end biosignal amplifier provides excellent signal resolution with 24-bit ADCs (analog-to-digital converters) for each channel for  precise measurement of evoked potentials

### **2.Types of Electrode Boxes**

-  64-channel electrode boxes : g.HEADbox - passive: For standard 1.5mm safety connectors used with conventional EEG electrodes
- One 16-channel box (g.HEADbox16): Used for additional inputs or external sensors

### **3.How was the impedance check done? How did you visually measure the impedance of the electrodes?  What is the purpose of an impedance check?**

- The impedance check was performed using the "Impedance Check" feature of g.tec Suite software, which allows real-time visualization of electrode impedances in both 2D and 3D topographic views. The process includes:
- Viewing a color-coded topographic map of the scalp where electrode impedance values are represented by different colors
- Identifying electrodes with poor impedance (non-green colors) and applying additional conductive gel to electrodes showing weak impedance
### **What is the purpose of an impedance check?**
- The impedance check ensures proper electrical contact between electrodes and scalp, which is critical for:
- Minimizing noise and artifacts in EEG recordings
- Ensuring consistent signal quality across all channels
- Reducing 50/60Hz power line interference

### **4. What are the types of biosignals acquired?**
This experiment specifically focused on:

- Auditory Evoked Potentials (AEPs): Electrical responses of the auditory pathways that occur within 10-15ms of an acoustic stimulus
- The system recorded these responses using EEG electrodes placed according to the international 10-20 system, with particular attention to electrodes positioned over temporal and central regions that best capture auditory processing
- The experiment also included the acquisition of other biosignals, such as: EOG and ECG for artifact rejection

```python
import matplotlib.pyplot as plt
import cv2
```

```python
def show_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #set the figure size
    plt.figure(figsize=(8, 6))
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.title("Raw Data")
    plt.show()
```

```python
show_image('eeg_plots/sub1_b1_data_raw_plot.png')
show_image('eeg_plots/sub1_b2_data_raw_plot.png')
show_image('eeg_plots/sub1_b3_data_raw_plot.png')
```
